Basically there are :
1. Numeric which includes integer,float and complex type .
2. Boolean 
3. Sequence type which includes string , list and tuples.
4. set 
5. Dictionary

Integer: It is a whole number which includes negative but not fractional numbers .
Float : It is a real number with floating point representaion.
complex:  a+bj where a is real and bj is imaginary part 

```python
a=-10.022
print(abs(a))
print(round(a,2))

num1=3+4j
num2=3-4j
print(num1.real)
print(num1.imag)
print(num1.conjugate())

print(num1+num2)
```

Type conversion: 
It is automatically done by python so ,

```python
a=5
print(a)
print(type(a))
b=float(a)
print(b)
print(type(b))
c=complex(a)
print(c)
print(type(c))
```

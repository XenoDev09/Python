```python
class Dog:
    def __init__(self,name,age): #constructor method
        self.name=name
        self.age=age
    
    def bark(self):
        print(f"{self.name} is barking!") 

my_dog=Dog("Buddy",3)

my_dog.bark()

print(my_dog.name)
print(my_dog.age)
```

PYTHON INHERITANCE:

```python
class Dog: #Main class
    def __init__(self,name,age):#Constructor
        self.name=name
        self.age=age

    def display_name(self):
        print(f"My dog's name is {self.name}")

class Labrador(Dog): # SingleInheritance
    def sound(self):
        print(f"{self.name}is barking!")

class GuideDog(Labrador): #Multi-Level Inheritance
    def guide(self):
        print("The bog is guiding!")

class Friendly:
    def greet(self):
        print("Friendly")

class GoldenRetriver(Dog,Friendly): #Multiple Inheritance
    def sound(self):
        print("GoldenRetriver barks!")

my_dog1=Dog("Nitu",23)#main class
my_dog1.display_name()

my_dog2=Labrador("Nischal",24)#single inheritance
my_dog2.display_name()
my_dog2.sound()

my_dog3=GuideDog("Sameer",23)#multilevel inheritace
my_dog3.guide()
my_dog3.sound()
my_dog3.display_name()

my_dog4=GoldenRetriver("Nischal",19)#multiple inheritance
my_dog4.sound()
my_dog4.greet()
my_dog4.display_name()
```

Conclusion: 
The constructor of main class is also used in the child class , and other child class can access the other parent class methods.

POLYMORPHISM:
1. Compile time polymorphism
2. Run time polymorphism

```python
class Calculator: #Method Overloading(Compile time)
    def add(self,a,b=0,c=0):
        return a+b+c
    
calc = Calculator()
print(calc.add(10))
print(calc.add(10,20))
print(calc.add(10,20,30))

class Dog:#Method Overriding(Run time)
    def sound(self):
        print("Its barking!")

class Labrador(Dog):
    def sound(self):
        print("Labrador is barking!")

dog1=Labrador()
dog1.sound()
```

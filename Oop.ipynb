PYTHON INHERITANCE:

```python
class Dog: #Main class
    def __init__(self,name,age):#Constructor
        self.name=name
        self.age=age

    def display_name(self):
        print(f"My dog's name is {self.name}")

class Labrador(Dog): # SingleInheritance
    def sound(self):
        print(f"{self.name}is barking!")

class GuideDog(Labrador): #Multi-Level Inheritance
    def guide(self):
        print("The bog is guiding!")

class Friendly:
    def greet(self):
        print("Friendly")

class GoldenRetriver(Dog,Friendly): #Multiple Inheritance
    def sound(self):
        print("GoldenRetriver barks!")

my_dog1=Dog("Nitu",23)#main class
my_dog1.display_name()

my_dog2=Labrador("Nischal",24)#single inheritance
my_dog2.display_name()
my_dog2.sound()

my_dog3=GuideDog("Sameer",23)#multilevel inheritace
my_dog3.guide()
my_dog3.sound()
my_dog3.display_name()

my_dog4=GoldenRetriver("Nischal",19)#multiple inheritance
my_dog4.sound()
my_dog4.greet()
my_dog4.display_name()
```

Conclusion: 
The constructor of main class is also used in the child class , and other child class can access the other parent class methods.

POLYMORPHISM:
1. Compile time polymorphism
2. Run time polymorphism

```python
class Calculator: #Method Overloading(Compile time)
    def add(self,a,b=0,c=0):
        return a+b+c
    
calc = Calculator()
print(calc.add(10))
print(calc.add(10,20))
print(calc.add(10,20,30))

class Dog:#Method Overriding(Run time)
    def sound(self):
        print("Its barking!")

class Labrador(Dog):
    def sound(self):
        print("Labrador is barking!")

dog1=Labrador()
dog1.sound()
```

```python
class Shape():
    def area(self):
        return "Undefined!"
    
class Rectangle(Shape):
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth

    def area(self):
        return self.length*self.breadth
    
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return 3.12*self.radius**2
    
shapes=[Rectangle(10,10),Circle(10)]
for shape in shapes:
    print(shape.area())
    
```

Encapsulation:

```python
class Dog:
    def __init__(self,name,breed,age):
        self.name=name
        self._breed=breed
        self.__age=age

    def public_method(self):
        print(f"{self.name} is of {self._breed} and is {self.__age} years old !")
    
    def _protected_method(self):
        print(f"{self.name} is of {self._breed} and is {self.__age} years old !")

    def __private_method(self):
        print(f"{self.name} is of {self._breed} and is {self.__age} years old !")

    def get_age(self):
        return self.__age
    
    def set_age(self,age):
        if age > 0:
            self.__age=age
        else:
            print("Invalid age!")

my_dog=Dog("bhunku","Golden",20)
print(f"My dog name is {my_dog.name}")
print(f"My dogs breed is {my_dog._breed}")

print(my_dog.get_age())
my_dog.set_age(25)
print(my_dog.get_age())
```

ABSTRACTION:

```python
from abc import ABC , abstractmethod #Importing Abstract Base Class along with abstract method

class Dog(ABC):
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    @abstractmethod
    def sound(self):
        pass

    @property    
    def display_name(self):
        pass

class Labrador(Dog):
    def sound(self):
        print(f"{self.name} is barking !")

    def display_name(self):
        print(f"{self.name} !")

dog=Labrador("hero",20)
dog.display_name()
dog.sound()
```

__init__ Method:

```python
class Person():
    def __init__(self,name,id_number):
        self.name=name
        self.id_number=id_number

    def display_detail(self):
        print(f"{self.name} has {self.id_number} id number !")
    
class Employee(Person):
    def __init__(self,name,id_number,salary,post):
        super().__init__(name,id_number)
        self.salary=salary
        self.post=post

    def display_detail(self):
        print(f"{self.name} has {self.id_number} and salary of {self.salary} in post {self.post}!")

p=Person("Nischal",10)
p.display_detail()
e=Employee("Manjeel",2,20000,"Manager")
e.display_detail()
```

Iterators in Python:
use of iter() and next() method :

```python
class EvenNumber:
    def __iter__(self):
        self.n=2
        return self
    def __next__(self):
        x=self.n
        self.n+=2
        return x
    
even=EvenNumber()
it=iter(even)
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
```

StopIteration Exception:

```python
li=[10,20,30]
it=iter(li)

while True:
    try:
        print(next(it))
    except StopIteration:
        print("The iteration is completed!")
        break
```

Python Exception Handling:

try:
   #code that might raise an error
except :
   #code that runs when error occurs
else:
  #code to run if no error occurs
finally:
  #code that runs anyway

```python
n=2
try:
    result=n/2
except ZeroDivisionError:
    print("It cannot be divisible by 0!")
```

```python
try:
    a,b=2,0
    c=a/b
except ZeroDivisionError:
    print("It cannot be divisible by 0!")
except ValueError:
    print("Its value error !")
finally:
    print("It cannot be divided!")
```

Python catching Errors:
1. Catching specific exceptions(Upper example)
2. Catching multiple exceptions
3. Catch all handlers and their risks

```python
# Catching multiple exceptions
a=["10","Nischal","20"]
try:
    c=int(a[0])+int(a[1])
except (ValueError,TypeError) as e: #This part is important
    print("Error",e)
except IndexError:
    print("It is out of index error !")
else:
    print("There is no error! ")
```

```python
# Catch all handlers and their risks
try:
    res="100"/20
except ArithmeticError:
    print("Its airthmetic error !")
except: #THIS PART IS IMPORTANT
    print("Something went wrong here !")
```

RAISE AN ERROR:
Syntax:
raise Exception_type("msg")

```python
def set(age):
    if age<=0:
        raise ValueError("Age cannot be smaller than 0")
try:
    set(-5)
except ValueError as e:
    print(e)
```

USER DEFINED EXCEPTION:
Creating our own exception and using it as exception!

```python
class InvalidAgeError(Exception):
    def __init__(self,age,msg="Age must be between 0 and 120 ",error_code=1001):
        self.age=age
        self.msg=msg
        self.error_code=error_code
        super().__init__(self.msg)
    def __str__(self):
        #return f"{self.age}->{self.msg}"
        return f"[Error_code:{self.error_code}] {self.age}->{self.msg}"
    
def set_age(age):
    if age<0 or age>120:
        raise InvalidAgeError(age)
    else:
        print(f"The age is set as {age}")

try:
    set_age(-5)
except InvalidAgeError as e:
    print(e)
finally:
    print("This is finally term !")
```

Standard Exceptions as a base class:

```python
class NetworkError(RuntimeError):
    def __init__(self,args):
        self.args=args
    
try:
    raise NetworkError("The network error is raised!")
except NetworkError as e:
    print(e)
```

```python
"""Scope of variables"""

def myfunc():
    a=10
    print("The value of a inside the function is:",a)

myfunc()
print(a) # This will give an error because 'a' is not defined outside the function

def myfunc2():
    global b
    b=20
    print("The value of b inside the function is:",b)

myfunc2()
print("The value of b outside the function is:",b)  # This will work because 'b' is declared as global
```

RETURN AND YEILD

```python
def fun():
    number=int(input("Enter a number:"))
    print("You entered:",number)
    return number
a=fun()
print("The returned value is:",a)

def fund():
    yield 1
    yield 2
    yield 3
for value in fund():
    print(value)
    
print(fund())
```

Global and non-local keywords

```python
a=10
b=20
def add():
    c=a+b
    print("The sum of these two numbers is {}".format(c))

add()

def outer():
    x=10
    def inner():
        nonlocal x
        x+=5
        print("Inner x:",x)
    inner()
    print("Outer x:",x)

outer()
```

FILE HANDLING:
Operations on file : file flow -->>
1. creating (At first checks as the file exits or not , if no then create the file)
2. opening 
3. reading
4. writing
5. closing

```python
with open("sample.txt","r+") as file:
    print(file.read())
    file.write("This is the second line! ")
    print(file.read())
    file.close()
```

```python
with open("sample.txt","a") as file:
    file.write("This is the third line !")
    file.close()
with open("sample.txt","r") as file:
    print(file.read())
```

There are 2 methods to use file operation :
They are :
1. use of with and as keywords(As like before)
2. Exception Handling

```python
try:
    file=open("sample.txt","r")
    print(file.read())
finally:
    file.close()
```

READING LINE BY LINE FILE :
1. for line in file:
   line.strip() 

2. Using readline()

```python
file=open("sample.txt","r")
for line in file:
    print(line.strip())
file.close()
```

```python
file=open("sample.txt","r")
line=file.readline()
while line:
    print(line.strip())
    line=file.readline()
file.close()
```

READING SPECIFIC PART OF A FILE:
Using bytes ,like file.read(10) means read the first 10 bytes!

```python
file=open("sample.txt","r")
print(file.read(10))
file.close()
```

READING CSV(COMMA SEPERATED VALUES) AND JSON(JAVASCRIPT OBJECT NOTATION) FILES:

```python
# CSV
import csv
with open("Test.csv",newline='') as csvfile:
    csvreader=csv.reader(csvfile)
    for row in csvreader:
        print(row)
```

```python
# JSON
import json
with open("Test.json","r") as jsonfile:
    data=json.load(jsonfile)
    print(data)
```

WRITING TO AN EXISTING FILE:
Using writelines()

```python
s=["This is the first line\n","This is second line\n","This is third line\n"]
with open("test.txt","w") as file:
    file.writelines(s)

file=open("test.txt","r")
print(file.read())
file.close()
```
